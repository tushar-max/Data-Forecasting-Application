```python
import numpy as np
import pandas as pd
```

```python
row = ['Country', 'Gender', 'Age-Group']
column = ['Year']
value = 'People'
def get_pt():
    df = pd.read_excel("Pivot Practice.xlsx",sheet_name="Sheet5")
    row = ['Country', 'Gender', 'Age-Group']
    column = ['Year']
    value = 'People'
    pivot_table = pd.pivot_table(df, values=value, index=row, columns=column, aggfunc='sum', fill_value=0, margins=True, margins_name='Total', observed=True)
    # Implement logic to fetch and process data
    # Return the data in JSON format
    return df,pivot_table
```

```python
def update_data(i,j,val,pivot_table,df):
    Tr = pivot_table.index[i]
    Tc = pivot_table.columns[j]
    mask = pd.Series(True, index=df.index)
    if isinstance(Tr,tuple):
        idx=0
        for i in Tr:
            mask &= (df[row[idx]]==i)
            idx+=1
    else:
         mask &= (df[row[0]]==Tr)
    if isinstance(column,tuple):
        idx=0
        for i in Tc:
            mask &= (df[column[idx]]==i)
            idx+=1
    else:
         mask &= (df[column[0]]==Tc)
    # print(mask)
    df.loc[mask,value] = val
    df.to_excel("Pivot Practice.xlsx",sheet_name="Sheet5",index=False)
    return df
```

```python
df,pivot_table = get_pt()
```

```python
df = update_data(0,0,20,pivot_table,df)
df.to_excel("Pivot Practice.xlsx",sheet_name="Sheet5",index=False)
```
